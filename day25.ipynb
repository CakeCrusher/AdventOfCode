```python
import copy
```

```python
def arrToStr(arr):
  return "".join(["".join(x) for x in arr])
def prettifyArr(arr):
  return "\n".join(["".join(x) for x in arr])
```

```python
east = ">"
south = ">"
stop = "."
```

```python

def prepareEx(ex):
  ex = ex.split("\n")
  return [list(x) for x in ex]

def move(bit, y, x, type, _ogEx, _newEx):
  ogEx = copy.deepcopy(_ogEx)
  newEx = copy.deepcopy(_newEx)
  dimX = len(ogEx[0]) - 1
  dimY = len(ogEx) - 1
  if (type == "east"):
    if (bit == ">"):
      moveX = x + 1
      if (moveX > dimX):
        moveX = 0
      if ogEx[y][moveX] == ".":
        # print("change x: ",newEx)
        newEx[y][moveX] = ">"
        newEx[y][x] = "."
  if (type == "south"):
    if (bit == "v"):
      moveY = y + 1
      if (moveY > dimY):
        moveY = 0
      if ogEx[moveY][x] == ".":
        # print("change y: ",newEx)
        newEx[moveY][x] = "v"
        # print(x,y)
        newEx[y][x] = "."
  return newEx

# print(move(".", 0, 0, "south", ex, ex))
```

```python
def run(type, nowEx):
  ogEx = copy.deepcopy(nowEx)
  newEx = copy.deepcopy(ogEx)
  for yidx, y in enumerate(ogEx):
    print("yidx: ", yidx)
    for xidx, x in enumerate(y):
      newEx = move(ogEx[yidx][xidx], yidx, xidx, type, ogEx, newEx)

  return newEx
# print(run("south", ex))
```

```python
def run_step(nowEx):
  newEx = copy.deepcopy(nowEx)
  newEx = run("east", newEx)
  newEx = run("south", newEx)
  nowExStr = arrToStr(nowEx)
  newExStr = arrToStr(newEx)
  if (nowExStr == newExStr):
    print("same")
    return "SAME"
  return newEx
# a = run_step(ex)
# b = run_step(a)
```
